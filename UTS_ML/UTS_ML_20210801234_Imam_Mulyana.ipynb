## Dataset

```python
import pandas as pd

data = {
    'hari': ['senin', 'selasa', 'rabu', 'kamis', 'jumat', 'sabtu', 'minggu'],
    'datang': [2, 3, 4, 1, 2, 5, 2],
    'biaya': [30000, 35000, 25000, 15000, 20000, 30000, 35000],
    'mahasiswa': ['ani', 'budi', 'jono', 'lono', 'joni', 'ani', 'budi']
}

df = pd.DataFrame(data)
df
```

## Menambahkan total biaya (mengalikan datang dan biaya)

```python
df['total_biaya'] = df['datang'] * df['biaya']
df
```

### a) Berapa rata-rata mahasiswa datang pada minggu ini?

```python
rata_rata_datang = df['datang'].mean()
print(f"Rata-rata mahasiswa datang pada minggu: {rata_rata_datang}")
```

### b) Kapan biaya tertinggi terjadi?



```python
hari_total_biaya_tertinggi = df.loc[df['total_biaya'].idxmax(), 'hari']
print(f"Total Biaya tertinggi pada hari: {hari_total_biaya_tertinggi}")

hari_biaya_tertinggi = df[df["biaya"] == df["biaya"].max()]
print(f"Biaya tertinggi pada hari: \n")
hari_biaya_tertinggi
```

### c) Hari apa biaya lebih dari 110000?

```python
hari_total_biaya_lebih_110000 =  df.loc[df['total_biaya'] > 110000, 'hari']
print(f"Hari total biaya lebih dari 110000: {hari_total_biaya_lebih_110000.tolist()}")
```

### d) Siapa yang paling banyak datang ke kampus?

```python
jumlah_datang = df.groupby('mahasiswa')['datang'].sum()
mahasiswa_terbanyak_datang = jumlah_datang.idxmax()
print(f"Mahasiswa yang paling banyak datang ke kampus: {mahasiswa_terbanyak_datang}")
```

### e) Siapa yang datang pada hari minggu?

```python
mahasiswa_hari_minggu = df[df['hari'] == 'minggu']['mahasiswa']

print(f"Mahasiswa yang datang pada hari Minggu: {mahasiswa_hari_minggu.tolist()}")
```

### f) Berapa biaya tertinggi dan terendah?

```python
biaya_tertinggi = df['biaya'].max()
biaya_terendah = df['biaya'].min()

print(f"Biaya tertinggi: {biaya_tertinggi}")
print(f"Biaya terendah: {biaya_terendah}")

total_biaya_tertinggi = df['total_biaya'].max()
total_biaya_terendah = df['total_biaya'].min()

print(f"Total Biaya tertinggi: {total_biaya_tertinggi}")
print(f"Total Biaya terendah: {total_biaya_terendah}")
```

### g) Berapa frekuensi datang tertinggi dan terendah?

```python
frekuensi_datang_tertinggi = df['datang'].max()
frekuensi_datang_terendah = df['datang'].min()

print(f"Frekuensi datang tertinggi: {frekuensi_datang_tertinggi}")
print(f"Frekuensi datang terendah: {frekuensi_datang_terendah}")
```

```python
from plotnine import ggplot, aes, geom_bar, labs, theme_minimal, theme, element_text
import pandas as pd

fakultas = ["Bisnis", "D3 Perhotelan", "ICT", "Ilmu Komunikasi", "Seni dan Desain"]
jumlah_mahasiswa = [260, 28, 284, 465, 735]
akreditasi = ["A", "A", "B", "A", "A"]

info_mahasiswa = pd.DataFrame({'Fakultas': fakultas, 'Jumlah Mahasiswa': jumlah_mahasiswa, 'Akreditasi': akreditasi})

gambar = (ggplot(info_mahasiswa, aes(x="Fakultas", y="Jumlah Mahasiswa", fill="Fakultas"))
          + geom_bar(stat="identity", width=1)
          + theme_minimal()
          + labs(title="Jumlah Mahasiswa Perfakultas", x="Fakultas", y="Jumlah Mahasiswa"))

print(gambar)
```
